```python
# Blood pressure
# the values recorded in Blood pressure are separated by a slash.
# the values recorded are systolic Bp and Diastolic Bp. Lets seperate the two values and convert the variable to numeric

def split_bp(x):
    if x==np.nan:
        return " "
    else:
        return x.split('/')
```
